```python
data_path = '../data/raw/games_best_sellers.csv'

df = pd.read_csv(data_path)

df.head()

df.info()

df.isna().sum()
```

```python
sns.histplot(data=df, x='price',kde=True)
plt.title('Distribuição de preço')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Serie temporal de preços ao longo do tempo')
sns.lineplot(data=df, x='release_date', y='price')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Jogos por ano')
df_lanc_ano = df['year_release'].value_counts()
sns.barplot(data=df_lanc_ano)
plt.show()
```

```python
df_lanc_mes = df['month_release'].value_counts()
plt.figure(figsize=(12,6))
plt.title('Jogos por mes')
sns.barplot(data=df_lanc_mes)
plt.show()
```

```python
df_preco_medio_ano = df.groupby(by='year_release',as_index=False).aggregate({'price':'mean'})
df_preco_medio_ano

#sns.lineplot(data=df_lanc_ano,x='year_release',y='price')
# %$
```

```python
plt.title('Preco medio por ano')
df_preco_medio_ano = df.groupby(by='year_release',as_index=False).aggregate({'price':'mean'})
sns.lineplot(data=df_preco_medio_ano,x='year_release',y='price',markers='o')
plt.show()
# %$
```

```python
plt.title('Preço e existencia de desconto')
sns.boxplot(data=df, x='has_discount',y='price')
plt.show()
```

```python
plt.title('Preço original x Preço com desconto')
sns.scatterplot(data=df, x='price_whitout_discount',y='price')
plt.show()
```

```python
plt.title('Matriz de correlação')
sns.heatmap(df.select_dtypes(exclude='str').corr(),cmap='coolwarm',annot=True)
plt.show()
```
